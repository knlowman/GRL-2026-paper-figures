### Calculate OHC for ensemble members

```python
# members: calculate OHC
for co2_scen in ['const','doub']:
    arr_name = f"{co2_scen}_ctrl_{start}_{end}_mem_list"
    for ens_idx in range(5):
        myVars[arr_name][ens_idx]["OHC"] = calc_OHC_anom(myVars[arr_name][ens_idx]["thetaoga"])
        
for co2_scen in ['const','doub']:
    for prof in profiles:
        for power in power_var_suff:
            arr_name = f"{co2_scen}_{prof}_{power}_{start}_{end}_mem_list"
            for ens_idx in range(5):
                myVars[arr_name][ens_idx]["OHC"] = calc_OHC_anom(myVars[arr_name][ens_idx]["thetaoga"])
```

```python
# members: calculate OHC differences
arr_name = f"doub_ctrl_{start}_{end}_diff_mem_list"
for ens_idx in range(5):
    myVars[arr_name][ens_idx]["OHC"] = calc_OHC_anom(myVars[arr_name][ens_idx]["thetaoga"])

for diff_type in ['const','1860','2xctrl','const_ctrl']:
    for prof in profiles:
        for power in power_var_suff:
            if diff_type == 'const':
                arr_name = f"const_{prof}_{power}_{start}_{end}_diff_mem_list"
            else:
                arr_name = f"doub_{prof}_{power}_{start}_{end}_diff_{diff_type}_mem_list"
            for ens_idx in range(5):
                myVars[arr_name][ens_idx]["OHC"] = calc_OHC_anom(myVars[arr_name][ens_idx]["thetaoga"])
```

### Calculate OHC for ensemble mean

```python
# mean: calculate OHC
for co2_scen in ['const','doub']:
    ds_name = f"{co2_scen}_ctrl_{start}_{end}"
    myVars[ds_name]["OHC"] = calc_OHC_anom(myVars[ds_name].thetaoga)
    for prof in profiles:
        for power in power_var_suff:
            ds_name = f"{co2_scen}_{prof}_{power}_{start}_{end}"
            myVars[ds_name]["OHC"] = calc_OHC_anom(myVars[ds_name].thetaoga)
```

```python
# mean: calculate OHC differences
for co2_scen in ['const','doub']:
    if co2_scen == 'doub':
        ds_name = f"{co2_scen}_ctrl_{start}_{end}_diff"
        myVars[ds_name]["OHC"] = calc_OHC_anom(myVars[ds_name].thetaoga)
    for prof in profiles:
        for power in power_var_suff:
            if co2_scen == 'doub':
                for diff_suff in ['const_ctrl','1860','2xctrl']:
                    ds_name = f"{co2_scen}_{prof}_{power}_{start}_{end}_diff_{diff_suff}"
                    myVars[ds_name]["OHC"] = calc_OHC_anom(myVars[ds_name].thetaoga)
            else:
                ds_name = f"{co2_scen}_{prof}_{power}_{start}_{end}_diff"
                myVars[ds_name]["OHC"] = calc_OHC_anom(myVars[ds_name].thetaoga)
```

```python
import os
import gc
import time as time_module
import dask

num_mem = 5
batch_size = 2   # try 2 first; increase to 3 or 5 if memory is okay

data_dir = "/vftmp/Kiera.Lowman/GRL_2026_data/single_members/"
os.makedirs(data_dir, exist_ok=True)

for diff_type in ["const-1860ctrl", "doub-2xctrl", "doub-1860exp", "doub-1860ctrl"]:
    for power in power_var_suff:
        for prof in profiles:

            ds_pref = get_ds_pref(diff_type, prof, power)
            ds_name = ds_pref + "_mem_list"

            print(f"\nWorking on {ds_name}")

            for i0 in range(0, num_mem, batch_size):
                i1 = min(i0 + batch_size, num_mem)
                idx_batch = list(range(i0, i1))

                print(f"Computing members {[i + 2 for i in idx_batch]}")
                t0 = time_module.time()

                # compute/load several members at once
                with dask.config.set(scheduler="threads"):
                    computed_batch = dask.compute(
                        *[myVars[ds_name][idx].load() for idx in idx_batch]
                    )

                print("BATCH COMPUTE seconds:", time_module.time() - t0)

                for idx, ds_mem in zip(idx_batch, computed_batch):
                    data_fname = f"{ds_pref}_mem{idx + 2}.nc"
                    full_path = os.path.join(data_dir, data_fname)

                    print(f"Writing {data_fname}")
                    t0 = time_module.time()

                    ds_mem.to_netcdf(
                        full_path,
                        mode="w",
                        format="NETCDF4",
                        engine="netcdf4",
                        encoding={
                            v: {"zlib": False, "chunksizes": None}
                            for v in ds_mem.data_vars
                        },
                    )

                    print("WRITE seconds:", time_module.time() - t0)

                for ds_mem in computed_batch:
                    ds_mem.close()

                del computed_batch
                gc.collect()
```
